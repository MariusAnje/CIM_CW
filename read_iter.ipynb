```python
import os
import numpy as np
parent_dir = "results/iter_mask"
file_list = os.listdir(parent_dir)
```

```python
line_name = {5:"S before",6:"acc before no noise",8:"acc after no noise",9:"S after",10:"S after recalc", 17:"mean acc"}
# (A,B) --> A splitter only one thing in a line if None || B True --> before A, False --> After A
line_type = {5:(":",False),6:(None,True),8:(" ", True),9:(":", False),10:(":", False),17:(":", False)}
line_content = {5:[],6:[],8:[],9:[],10:[],17:[]}

for fn in file_list:
    with open(os.path.join(parent_dir,fn),"r") as f:
        content = f.read().split("\n")
    for key in line_name.keys():
        splitter, before = line_type[key]
        line = content[key].split(splitter)
        if before:
            value = float(line[0])
        else:
            value = float(line[1])
        line_content[key].append(value)
```

```python
for key in line_name.keys():
    print(f"{line_name[key]:20s}: {np.mean(line_content[key])}")
```
